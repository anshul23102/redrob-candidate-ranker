```python
# Convert the bundled 50-candidate sample (JSON list) to JSONL — the ranker's input format
import json
cands = json.load(open('data/sample_candidates.json'))
with open('sample.jsonl','w') as f:
    for c in cands:
        f.write(json.dumps(c) + '\n')
print(f'{len(cands)} candidates ready')
```

```python
# Rank them — the exact pipeline that produced our submission (scoring, integrity,
# semantic artifact lookup, grounded reasoning). No model download, no network.
!python rank.py --candidates ./sample.jsonl --out ./sandbox_ranking.csv --topk 50
```
